```python
import csv
from dataclasses import dataclass, asdict

@dataclass
class Stock:
    name: str
    symbol: str
    price: float

class StockManager:
    def __init__(self):
        self.stocks = []

    def add_stock(self, stock: Stock):
        self.stocks.append(stock)

    def add_bulk(self, stock_list):
        for stock in stock_list:
            self.add_stock(stock)

    def export_to_csv(self, filename='stocks.csv'):
        with open(filename, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["Name", "Symbol", "Price"])
            writer.writeheader()
            for stock in self.stocks:
                writer.writerow({
                    "Name": stock.name,
                    "Symbol": stock.symbol,
                    "Price": stock.price
                })
        print(f"[INFO] Exported {len(self.stocks)} stocks to {filename}")
```

```python
import csv
import pandas as pd
from dataclasses import dataclass, asdict

@dataclass
class Car:
    brand: str
    model: str
    price: float

class CarManager:
    def __init__(self):
        self.cars = []

    def add_car(self, car: Car):
        self.cars.append(car)

    def add_bulk(self, car_list):
        for car in car_list:
            self.add_car(car)

    def export_to_csv(self, filename="car_data.csv"):
        with open(filename, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["Brand", "Model", "Price"])
            writer.writeheader()
            for car in self.cars:
                writer.writerow({
                    "Brand": car.brand,
                    "Model": car.model,
                    "Price": car.price
                })
        print(f"[INFO] Exported {len(self.cars)} cars to {filename}")

    def import_from_csv(self, filename="car_data.csv"):
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                car = Car(row['Brand'], row['Model'], float(row['Price']))
                self.add_car(car)
        print(f"[INFO] Imported cars from {filename}")

    def to_dataframe(self):
        return pd.DataFrame([asdict(c) for c in self.cars])
```
