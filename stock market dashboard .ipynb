```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('stocks.csv')

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Functions for visualizations

def plot_stock_price(name):
    # Ensure correct column name
    name_data = df[df['Name'] == name]
    
    if name_data.empty:
        print(" Stock not found!")
        return
    
    # Plot the Price over time (since only price is available, we'll just plot it by index)
    plt.figure(figsize=(12, 6))
    plt.plot(name_data.index, name_data['Price'], label='Price', marker='o')
    plt.title(f"{name} Stock Price Over Time")
    plt.xlabel("Index")
    plt.ylabel("Price ($)")
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def plot_all_stock_prices():
    # Plot the Price for all stocks in the dataset
    plt.figure(figsize=(12, 6))
    for name in df['Name'].unique():
        name_data = df[df['Name'] == name]
        plt.plot(name_data.index, name_data['Price'], label=name)
    plt.title("Stock Prices for All Companies")
    plt.xlabel("Index")
    plt.ylabel("Price ($)")
    plt.grid(True)
    plt.legend(loc='best')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Sample usage for interactive dashboard
while True:
    print("\nOptions:")
    print("1. Plot stock price of a specific company")
    print("2. Plot stock prices of all companies")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        name = input("Enter stock name: ")
        plot_stock_price(name)
    elif choice == '2':
        plot_all_stock_prices()
    elif choice == '3':
        print("Exiting dashboard...")
        break
    else:
        print(" Invalid choice, try again.")
```
